```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given Data
X = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
y = np.array([30, 45, 51, 57, 60, 65, 70, 71])

# Step 1: Compute Mean of X and y
X_mean = np.mean(X)
y_mean = np.mean(y)

# Step 2: Compute Slope (m)
m = np.sum((X - X_mean) * (y - y_mean)) / np.sum((X - X_mean) ** 2)

# Step 3: Compute Intercept (b)
b = y_mean - m * X_mean

# Print Computed Values
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")

# Step 4: Fit Using Scikit-Learn
regressor = LinearRegression()
regressor.fit(X, y)

# Step 5: Compare Results
print(f"Scikit-Learn Slope (m): {regressor.coef_[0]}")
print(f"Scikit-Learn Intercept (b): {regressor.intercept_}")

# Compute Partial Derivatives (Gradients)
n = len(y)  # Number of data points
y_pred = regressor.predict(X)  # Predicted values

# Compute gradients
dm = - (1/n) * np.sum(X * (y - y_pred))  # Partial derivative w.r.t. m
db = - (1/n) * np.sum(y - y_pred)  # Partial derivative w.r.t. b

print(f"Partial Derivative with respect to m: {dm}")
print(f"Partial Derivative with respect to b: {db}")


# Assuming 'regressor' is your trained LinearRegression model
y_pred = regressor.predict(X)  # Get predictions
errors = y_pred - y  # Calculate errors
print(f"Errors: {errors}")
squared_errors = errors ** 2  # Square errors
print(f"Squared Errors:{squared_errors}")
mse = np.mean(squared_errors)  # Calculate mean

print(f"Mean Squared Error (Cost Function): {mse/2}")

# Step 6: Plot Regression Line
plt.scatter(X, y, color='red', label="Actual Data")
plt.plot(X, regressor.predict(X), color='blue', label="Regression Line")
plt.xlabel("X (Input)")
plt.ylabel("y (Target)")
plt.legend()
plt.show()
```
